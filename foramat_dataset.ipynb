```python
def prompt_no_input(row):
    return ("Ci-dessous se trouve une instruction qui décrit une tâche. "
            "Écrivez une réponse qui complète de manière appropriée la demande.\n\n"
            "### Instruction :\n{instruction}\n\n### Réponse :\n").format_map(row)


def prompt_input(row):
    return ("Ci-dessous se trouve une instruction qui décrit une tâche, associée à une entrée qui fournit un contexte supplémentaire. "
            "Écrivez une réponse qui complète de manière appropriée la demande.\n\n"
            "### Instruction :\n{instruction}\n\n### Entrée :\n{ saisir}\n\n### Réponse :\n").format_map(row)

def map_row(row):
  if row[' saisir'] == '':
    new_col = prompt_no_input(row)
  else:
    new_col = prompt_input(row)

  row['prompt'] = new_col
  return row

EOS_TOKEN = "<|endoftext|>"

def map_output(row):
  row['completion']  = row[' sortir'] + EOS_TOKEN
  return row
```

```python
data = load_dataset("tbboukhari/Alpaca_french_instruct")

data = data.map(map_row)
data = data.map(map_output)

data = data.remove_columns(['instruction', " saisir", ' sortir'])


data = data['train'].train_test_split(test_size = 0.05)

```
